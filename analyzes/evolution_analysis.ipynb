```python
from datetime import datetime, timedelta
from os.path import expanduser
import pandas as pd
import numpy as np

cats = ["Alt-right", "Alt-lite", "I.D.W.", "Media"]
names = ["Alt-right", "Alt-lite", "IDW", "control"]
colors = {"Alt-right": "#f93e49", "Alt-lite": "#F29F05", "I.D.W.": "#019441", "Media": "#35598a"}
bins_like_name = ["0", "1", "2-5", "6-100", "100"]
x_label = [str(i) for i in range(1, 11)]
years = ["2016", "2017", "2018"]
lines = ['-', '--', '-.', ':']
df_path = "./../data/sentiment/dataframes/"
img_path = "./../imgs/"
```

```python
def human_format(num, pos=None):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.2f%s' % (num, ['', 'K', 'M', 'B', 'T', 'P'][magnitude])
```

```python
attributes = ['SEVERE_TOXICITY']
persp_df_path = df_path + "perspective_df/evolution/"
perspective = []

fig = plt.figure(
    constrained_layout=False,
    figsize=(4.5, 2.5),
)
gs = fig.add_gridspec(nrows=1, ncols=1, wspace=0.2, hspace=0.45, top=0.90)

# adds subplots
ax1 = fig.add_subplot(gs[0, 0])

axs = [ax1]

# Plots cdfs on the top plots
for ax, year in zip(axs, range(len(years))):
    for emotion in attributes:
        for i in range(len(names)):
            perspective = pd.read_csv(f"{persp_df_path}{names[i]}_perspective_evolution10.csv")
            x = perspective.year
            y = perspective[emotion]
            y_dyd = perspective[emotion + "_dyd"]
            y_dyu = perspective[emotion + "_dyu"]

            ax.plot(x, y, lines[i], color=colors[cats[i]])
            ax.fill_between(x, y_dyd,y_dyu, alpha=0.1, color=colors[cats[i]])

    
    # Sets ticks, titles and limits
    ax.set_title("Authors with 10+ comments", fontsize=15)
    ax.set_xlim(1, 10)
    
for ax in axs[1:]:
    ax.yaxis.set_tick_params(labelbottom=False)
    
# Adds legend on the top
leg = ax1.legend(cats,
           loc='upper center',
           bbox_to_anchor=(0.5, 1.5),
           ncol=2,
           fancybox=False,
           shadow=False,
           frameon=False,
           edgecolor=None,
           fontsize=13)


for line in leg.get_lines():
    line.set_linewidth(3.0)

# Common axis config, grids, etc
for ax in axs:
    ax.xaxis.grid(color='gray', linestyle='solid')
    ax.yaxis.grid(color='gray', linestyle='dashed')
    ax.xaxis.set_tick_params(rotation=0)
    ax.set_xlabel("")
    ax.set_ylabel("")

# Adds labels
_ = ax1.set_ylabel("Severe Toxicity")
_ = ax1.set_xlabel("Comment Number")

# Saves figs
plt.savefig(f"{img_path}perspective/evolution/perspective_radicalization_10.pdf", bbox_inches = 'tight')       
plt.savefig(f"{img_path}perspective/evolution/perspective_radicalization_10.png", bbox_inches = 'tight')       
```
